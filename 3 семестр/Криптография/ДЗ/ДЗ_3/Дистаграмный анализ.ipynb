# Частотный анализ исторических шифров.

## Шифр Атбаш

Построение частотный гистограммы `шифра Атбаша`

```python
import matplotlib.pyplot as plt
import string
from collections import Counter
import itertools

RUSSIAN_ALP="абвгдежзийклмнопрстуфхцчшщъыьэюя"

def create_histogram(text: str, x: int, lab: str, alp=RUSSIAN_ALP, count_col=2) -> None:
    if ' ' in text:
        alphabet = list(alp.split())
        counter = Counter(text.lower().split())
    else:
        alphabet = list(alp)
        counter = Counter(text.lower())
    frequencies = [counter[letter] for letter in alphabet]
    plt.figure(figsize=(15,8))
    plt.subplot(count_col, 1, x)
    plt.bar(alphabet, frequencies)
    plt.xlabel('Буква')
    plt.ylabel('Частота')
    plt.title(lab)
    plt.show()

create_histogram("НевсетеповараЗПТчтосдлинныминожамиходятТЧК", 1, "Открытый текст")

create_histogram("тъэоънърсэяпяшрнинсксыаноыфчттдучтсщяучних", 2, "Шифр Атбаш")

```

Размытия статистических характеристик в данном шифре не происходит, т.к. буквы алфавита только лишь меняются ролями. В итоге использования данного шифра получаем зеркальное отражение статистических характеристик в виду особенности алгоритма замены.

## Квадрат Полибия

```python
create_histogram("НевсетеповараЗПТчтосдлинныминожамиходятТЧК", 1, "Открытый текст", RUSSIAN_ALP, 3)

create_histogram("32 16 13 36 16 41 16 34 33 13 11 35 11 22 34 41 46 41 33 36 15 26 23 32 54 31 23 32 33 21 11 31 23 44 33 15 62 41 41 46 25", 3, "Квадрат Полибия", "11 12 13 14 15 16 21 22 23 24 25 26 31 32 33 34 35 36 41 42 43 44 45 46 51 52 53 54 55 56 61 62", 3)
```

```python
create_histogram("НевсетеповараЗПТчтосдлинныминожамиходятТЧК", 1, "Открытый текст")

alp = str(' '.join(((''.join(x) for x in itertools.product("0123456789", repeat=2)))))

create_histogram("01 50 30 03 90 53 05 10 98 02 10 56 06 00 38 02 30 60 02 60 36 05 30 35 09 01 00 00 07 05 90 47 03 30 48 07 10 14 08 70 72 02 70 56 06 80 35 02 40 58 01 90 72 06 80 12 09 10 70 01 90 82 07 6", 2, "Матричный шифр", alp)
```

## Шифр Плейфера

```python
create_histogram("НевсетеповараЗПТчтосдлинныминожамиходятТЧК", 1, "Открытый текст")

alp = itertools.product(RUSSIAN_ALP, repeat=2)
tmp = []
for i in alp:
    tmp.append(''.join(i))
en_message = "не вс ет еп ов ар аз пт чт ос дл ин ны ми но жа ми хо дя тф тч кф"

alp = list(filter(lambda x: x in en_message, tmp))
alp = ' '.join(alp)

create_histogram("не вс ет еп ов ар аз пт чт ос дл ин ны ми но жа ми хо дя тф тч кф", 2, "Шифр Плейфера", alp)
create_histogram("раатвхжмвиензсицшснтлфзобчпкояавпктртбухсшлу", 2, "Шифр Плейфера")
```

Шифр Плейфера является шифром биграммной замены, поэтому если анализировать его биграммы на данном коротком открытом тексте, гистограмма асболютно ровная
Если анализовать его по одному символу, то шифр размывает статистические характеристики, но присутствуют "пики"

## Одноразовый блокнот Шеннона

ГЕНЕРАТОР ГАММЫ


```python
create_histogram("НевсетеповараЗПТчтосдлинныминожамиходятТЧК", 1, "Открытый текст")

alp = list(map(str, list(range(1, 32))))
for i in range(len(alp)):
    while len(alp[i]) != 2:
        alp[i] = '0' + alp[i]
alp =  ' '.join(alp)
create_histogram("15 26 14 08 11 27 21 26 24 31 20 15 14 24 07 05 09 23 10 24 26 04 08 08 07 08 16 23 11 15 14 03 14 29 01 05 10 08 02 29 01 07", 2, "одноразовый блокнот шеннона", alp)
```

## Шифр ElGamal

```python
create_histogram("НевсетеповараЗПТчтосдлинныминожамиходятТЧК", 1, "Открытый текст")
en_message = "3656 3624 3612 3672 5712 3603 3624 5732 3660 5706 3604 3668 3604 3632 3664 5738 3623 5738 3660 5736 3620 3648 5718 3656 3656 3639 3652 3636 5657 3036 2836 0436 5257 1836 1536 6036 2057 6457 3836 0357 4857 22"
alp = list(map(str, list(range(1, 10000))))
for i in range(len(alp)):
    while len(alp[i]) != 4:
        alp[i] = '0' + alp[i]
alp = ' '.join(list(filter(lambda x: x in en_message, alp)))

create_histogram(en_message, 2, "ELGAMAL", alp)
```

Шифр ElGamal показывает наилучшее размытие статистических характеристик. График практически ровный.
